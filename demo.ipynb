```python
import os
```

```python
base_dir = 'poses'
```

```python
filenames = os.listdir(base_dir)
for filename in filenames:
    if not filename.startswith('.'):
        print('filename:', filename)
        filename = os.path.join(base_dir, filename)
        cmd = 'python yogaPoser.py -i '
        os.system(cmd+filename.replace(' ', '\ '))
```

```python
cmd = 'python yogaPoser.py -v '
filename = '15-yoga-poses.m4v'
os.system(cmd+filename.replace(' ', '\ '))
```
