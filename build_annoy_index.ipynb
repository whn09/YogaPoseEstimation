```python
import pandas as pd
from annoy import AnnoyIndex
```

```python
dataset = pd.read_csv('yoga/data.csv', names=['label']+['f'+str(i) for i in range(8)])
```

```python
poses = ['boundAngle', 'catCow', 'chairPose', 'cobra', 'downwardDog',
       'forwardBend', 'garland', 'halfForwardBend', 'highLunge', 'plank',
       'reverseWarrior', 'seatedSpinalTwist', 'threeLegDownwardDog',
       'tree', 'triangle']
print('poses:', len(poses))
```

```python
dataset['label'] = dataset['label'].apply(lambda x: poses.index(x))
```

```python
dataset
```

```python
keys = []
embs = []
def func(x):
    key = int(x['label'])
    emb = [float(x['f'+str(i)]) for i in range(8)]
    keys.append(key)
    embs.append(emb)
dataset.apply(lambda x: func(x), axis=1)
```

```python
keys
```

```python
embs
```

```python
import pickle

f = 8
u = AnnoyIndex(f, 'angular')  # Length of item vector that will be indexed
for key, value in enumerate(embs):
    u.add_item(key, value)
u.build(100) # 100 trees
u.save('yoga/helpers/model/index_advance.ann')
pickle.dump(keys, open('yoga/helpers/model/keys.pkl', 'wb'))
```
